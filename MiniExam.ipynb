````python
def gen_exams(ID):
    import math
    a = int(ID[-1])%6
    b = int(ID[-1])%3
    print(f'รหัสนักศึกษา {ID} ทำข้อ {a} และ {b if a != b else 5-b}')
````

Merge ตาราง

````python
flights_airports = flights.merge(airports,how='left',left_on='รหัสสนามบินปลายทาง',right_on='รหัสสนามบิน') 
flights_airports
````

ข้อ 0

````python
flights_airports[['รหัสสนามบินปลายทาง','รัฐ']]
````

````python
Top_import = flights_airports.groupby('รัฐ').count().sort_values(by=['รหัสสนามบินปลายทาง'],ascending=False) #หารัฐที่มีสายการบินขาเข้ามากที่สุด
Top_import.iloc[:1,4:5]
````

````python
Lowest_import = flights_airports.groupby('รัฐ').count().sort_values(by=['รหัสสนามบินปลายทาง'],ascending=True) #หารัฐที่มีสายการบินขาเข้าน้อยที่สุด
Lowest_import.iloc[:1,4:5]
````

-CA คือรัฐที่มีสายการบินขาเข้ามากที่สุด

-AK คือรัฐที่มีสายการบินขาเข้าน้อยที่สุด

ข้อ 3

````python
data_flights2 = flights.groupby('สายการบิน')[['วันในเดือน']].count()
data_flights2
# จัดกลุ่มด้วย 'สายการบิน' และ count
````

````python
data_flights2.reset_index(inplace=True)
data_flights2
# reset index เพื่อเอา index ไปใช้งานต่อ
````

````python
data_flights2_sort = data_flights2.sort_values(by=['วันในเดือน'],ascending=False)
data_flights2_sort
# เรียงลำดับจากมากไปน้อย ตามจำนวนเที่ยวบิน ดูจากวันในเดือน
````

````python
data_flights2_sort_max = data_flights2.sort_values(by=['วันในเดือน'],ascending=False)
data_flights2_sort_max.iloc[:1,:]
# สายการบินที่มีเที่ยวบินมากที่สุด
````

````python
data_flights2_sort_min = data_flights2.sort_values(by=['วันในเดือน'],ascending=False)
data_flights2_sort_min.iloc[-1:,:]
# สายการบินที่มีเที่ยวบินน้อยที่สุด
````

- สายการบินที่มีเที่ยวบินมากที่สุดคือ WN 

- สายการบินที่มีเที่ยวบินน้อยที่สุดคือ HA
